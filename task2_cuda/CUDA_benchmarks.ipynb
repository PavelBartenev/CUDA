### KERNEL SUM

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(font_scale=1.3, style='whitegrid', palette='Set2')
```

```python
file = open('cuda_sum_time')
sum_times = np.loadtxt(file)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(sum_times)
plt.xlabel('Количество потоков')
plt.ylabel('Время работы ядра, сек')
plt.title('Kernel sum, Vector_size=$2^{20}$')
plt.xlim(-1, 100)
plt.show()
```

```python
file = open('cuda_sum_time_vecsize')
sum_times_vecsize = np.loadtxt(file)
vecsizes = np.arange(1, 2**20, 100);
```

```python
plt.figure(figsize=(15, 7))
plt.plot(vecsizes, sum_times_vecsize)
plt.xlabel('Размер вектора')
plt.ylabel('Время работы ядра, сек')
plt.title('Kernel sum, ThreadsPerBlock=$256$')
plt.ylim(0.005, 0.04)
plt.show()
```

## Scalar Mul
